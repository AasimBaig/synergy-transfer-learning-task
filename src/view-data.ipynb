```python
import os
import PIL
import glob
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
DATA_DIR = "/home/aasim/synergy-ai-task/data/"
train_data = os.path.join(DATA_DIR, "train")
val_data = os.path.join(DATA_DIR, "val")
```

```python
subfolders = [f.path for f in os.scandir(train_data) if f.is_dir()]
subfolders
```

```python
image_paths = []
for x in subfolders:
    image_paths.extend(glob.glob(os.path.join(x, "*.jpg")))
    target = [x.split("/")[-2] for x in image_paths]
```

```python
lb_enc = LabelEncoder()
target_enc = lb_enc.fit_transform(target)
target_enc = np.array(target_enc)

df2 = pd.DataFrame({"image_paths":image_paths[:], "targets":target_enc[:]}).sample(frac=1).reset_index(drop=True)
df2.to_csv("train_data.csv", index=False)
```
